```python
# Wright Fisher model
import matplotlib.pyplot as mp
import numpy as np
import os

FOLDER = "/mnt/primevo/work/timothe_dandoy/SLiM_Project/tmp/WF_NWF/"
WF_Pi_ratio = []
WF_PiA = []
WF_PiX = []
WF_PiY = []

NWF_Pi_ratio = []
NWF_PiA = []
NWF_PiX = []
NWF_PiY = []

for NUM in range(1,1001):
    File = FOLDER+"Pi_values"+str(NUM)+".txt"
    if os.path.isfile(File):
        Text = open(File, 'r')
        Values = Text.read().split()
        WF_Pi_ratio.append(float(Values[1])/float(Values[0]))
        WF_PiA.append(float(Values[0]))
        WF_PiX.append(float(Values[1]))
        WF_PiY.append(float(Values[2]))
        Text.close()

    File = FOLDER+"NWF_Pi_values"+str(NUM)+".txt"
    if os.path.isfile(File):
        Text = open(File, 'r')
        Values = Text.read().split()
        NWF_Pi_ratio.append(float(Values[1])/float(Values[0]))
        NWF_PiA.append(float(Values[0]))
        NWF_PiX.append(float(Values[1]))
        NWF_PiY.append(float(Values[2]))
        Text.close()
    

WF_mean = np.mean(WF_Pi_ratio)
NWF_mean = np.mean(NWF_Pi_ratio)
legend = ["PiA","PiX","PiY"]

print(np.mean(WF_PiA))
print("Number of WF simulations: ",len(WF_Pi_ratio))
print("WF Average Ratio: ",WF_mean)
print()
print(np.mean(NWF_PiA))
print("Number of NWF simulations: ",len(WF_Pi_ratio))
print("NWF Average Ratio: ",NWF_mean)


print("------------------------------------------------------------------------------------------------------------")
mp.subplot(1,2,1)
mp.boxplot([WF_Pi_ratio],tick_labels=["PiX / PiA"],showmeans=True)
mp.title('WF')

mp.subplot(1,2,2)
mp.boxplot([NWF_Pi_ratio],tick_labels=["PiX / PiA"],showmeans=True)
mp.title('NWF')
mp.show()

print("------------------------------------------------------------------------------------------------------------")
mp.subplot(1,2,1)
mp.boxplot([WF_PiA,WF_PiX,WF_PiY],tick_labels=legend,showmeans=True)
mp.title('WF')

mp.subplot(1,2,2)
mp.boxplot([NWF_PiA,NWF_PiX,NWF_PiY],tick_labels=legend,showmeans=True)
mp.title('NWF')
mp.show()
```
