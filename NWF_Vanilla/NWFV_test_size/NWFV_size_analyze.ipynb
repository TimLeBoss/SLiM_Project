```python
# Pi Values
import matplotlib.pyplot as mp
import numpy as np
import os 

FOLDER = "/mnt/primevo/work/timothe_dandoy/SLiM_Project/tmp/NWFV_test_size/"
Lm = [10,20,30,40,50,60,70]
LK = [250,500,750,1000,1250,1500]
HeatMap = np.zeros((len(Lm),len(LK)))

for i in range(len(Lm)):
    m = Lm[i]
    for j in range(len(LK)):
        K = LK[j]
        Pi_A = []
        for NUM in range(1,501):
            File = FOLDER + f"NWFV_size_K_{K}_m_{m}_{NUM}.txt"
            if os.path.isfile(File):
                Text = open(File, 'r')
                Values = Text.read().split()
                Pi_A.append(float(Values[0]))
                Text.close()       
        Pi_median = np.median(Pi_A)
        HeatMap[i,j] = Pi_median

        print(m,K)
        print(len(Pi_A)," simulations")
        print()


fig, ax = mp.subplots(figsize=(20, 6))

# Affichage des valeurs dans une grille (sans imshow)
for i in range(len(Lm)):
    for j in range(len(LK)):
        value = HeatMap[i][j]
        formatted = f"{value:.2e}"  # 2 chiffres significatifs, notation scientifique
        ax.text(j, i, formatted, ha='center', va='center', color='black')

# Dfinir les axes
ax.set_xticks(range(len(LK)))
ax.set_xticklabels(LK, rotation=45, ha='right', rotation_mode='anchor')

ax.set_yticks(range(len(Lm)))
ax.set_yticklabels(Lm)

# Ajouter les labels
ax.set_xlabel("K")
ax.set_ylabel("m")
ax.set_title("PiA")

# Optionnel : ajouter une grille pour mieux visualiser la "structure"
ax.set_xlim(-0.5, len(LK) - 0.5)
ax.set_ylim(-0.5, len(Lm) - 0.5)
ax.invert_yaxis()  # pour que l'origine soit en haut comme dans une heatmap

# Activer la grille
#ax.grid(True, color='lightgray', linestyle='-', linewidth=0.5)

# Affichage
mp.tight_layout()
mp.show()
```
