```python
#Number of offsprings
import matplotlib.pyplot as mp
import numpy as np
import os 
from collections import Counter
import pandas as pd
from scipy.stats import multinomial
from IPython.display import display

#------------------------------------------------------------------------------------------------------------------
## Fonctions
#------------------------------------------------------------------------------------------------------------------

def Multinomial (n,p):
    """
    Fonction qui simule la realisation d'une d'une loi multinomial de parametres n et p 
    """
    Sortie = [0 for j in range (len(p))]
    for k in range(n):
        u = random.random()
        i = 0
        sp = p[i]
        while u >= sp:
            i += 1
            sp += p[i]
        Sortie [i] += 1
    return Sortie     

def M(r,r_bar,R,N):
    S = 0
    for i in range(N):
        S+= (r[i] - r_bar[i])**2
    return (N/R**2)*S

#------------------------------------------------------------------------------------------------------------------
## Main Program
#------------------------------------------------------------------------------------------------------------------
FOLDER = "/mnt/primevo/work/timothe_dandoy/SLiM_Project/tmp/NWF_LifeTable/NWF_AoD_NOff_AD/"
#FOLDER_output = "/mnt/primevo/work/timothe_dandoy/SLiM_Project/done/NWF_LifeTable/NWF_AoD_NOff_AD/"

Models = ["All_Same","Multinomial","1_Reproducer"]
Data = pd.DataFrame({})
Nind = 5000
Rtot = 10000
Ngen = 2

for mod in Models:
    rf = []
    rm = []
    tf = []
    tm = []
    #Values collection
    if mod == "All_Same":
        rf = [Rtot/Nind for i in range(Nind)]
        rm = [Rtot/Nind for i in range(Nind)]
        tf = [Ngen for i in range(Nind)]
        tm = [Ngen for i in range(Nind)]

    elif mod == "Multinomial":
        rf = multinomial.rvs(Rtot, [1/Nind for i in range(Nind)])
        rm = multinomial.rvs(Rtot, [1/Nind for i in range(Nind)])
        tf = [Ngen for i in range(Nind)]
        tm = [Ngen for i in range(Nind)]

    elif mod == "1_Reproducer":
        rf = [Rtot]+[0 for i in range(Nind-1)]
        rm = [Rtot]+[0 for i in range(Nind-1)]
        tf = [Ngen for i in range(Nind)]
        tm = [Ngen for i in range(Nind)]
    
    # Males Reproductive Skew
    Rm = sum(rm)
    Tm = sum(tm) 
    Nm = len(rm)
    rm_bar = []
    for ti in tm:
        rm_bar.append(ti*Rm/Tm)
    Mm_observed = M(rm,rm_bar,Rm,Nm)

    Mm_multi = 0
    for j in range (100):
        rm_multi = multinomial.rvs(Rm, [ti/Tm for ti in tm])
        Mm_multi += M(rm_multi,rm_bar,Rm,Nm)
    RSm = Mm_observed - Mm_multi/100
    
    # Females Reproductive Skew
    Rf = sum(rf)
    Tf = sum(tf) 
    Nf = len(rf)
    rf_bar = []
    for ti in tf:
        rf_bar.append(ti*Rf/Tf)
    Mf_observed = M(rf,rf_bar,Rf,Nf)
    Mf_multi = 0
    for j in range (100):
        rf_multi = multinomial.rvs(Rf, [ti/Tf for ti in tf])
        Mf_multi += M(rf_multi,rf_bar,Rf,Nf)
    RSf = Mf_observed - Mf_multi/100 

    Model = pd.DataFrame([{"Males": RSm, 'Females': RSf}],index=[mod])
    Data = pd.concat([Data, Model])

#Ploting
Title = "<span style='font-size:20px; font-weight:bold;'>Reproductive Skew of males and females for border cases</span>"
display (
    Data.style\
        .set_table_attributes("style='width:100%; font-size:15px'")\
        .set_properties(**{'text-align': 'left'})\
        .set_table_styles([
        {'selector': 'th.col_heading', 'props': [('text-align', 'left')]},   # colonnes
        {'selector': 'th.row_heading', 'props': [('text-align', 'left')]}    # index (lignes)
        ])\
        .set_caption(Title)
)
```
