```python
# Star Migration
import matplotlib.pyplot as mp
import numpy as np
import os 

print("Star Migration")
print()

FOLDER = "/mnt/primevo/work/timothe_dandoy/SLiM_Project/tmp/Test_Mig_subpop_NWFV/"
expected_ratio = 0.75
expected_teta = round(4*500*1.25e-8,8)
LPro = [0.0]+[k/20 for k in range(1,20)]+[1.0]

Pi_Sp = []
for pop in range (1,11):
    Pi_Sp.append([0 for k in range(21)])
compteur = 0
#PIA = [0 for k in range(21)]
for ProbMig in LPro:
    Pi_ratio = [[] for pop in range(1,11)]
    
    for NUM in range(1,501):
        File = FOLDER+"NWFV_Subpop_star_Migration"+str(ProbMig)+","+str(NUM)+".txt"
        if os.path.isfile(File):
            Text = open(File, 'r')
            Values = Text.read().split()
            if len(Values) == 30:
                for val in range (0,28,3):
                    Pi_ratio[int(val/3)].append(float(Values[val+1])/float(Values[val]))
                    #PiA.append(float(Values[0]))
            Text.close()
    for pop in range(0,10):
        Pi_Sp[pop][compteur] = Pi_ratio[pop]     
        #PIA [compteur] = PiA

    compteur += 1


Labels = [str(k/20) for k in range(21)]
for pop in range (1,11):
    print("------------------------------------------------------------------------------------------------------------")
    print('Subpopulation ',pop)
    Values = Pi_Sp[pop-1]
    
    fig, ax = mp.subplots(figsize=(10, 5))
    ax.boxplot(Values,tick_labels=Labels)
    ax.set_title("Star_Migration")
    ax.set_xlabel("Males Migration Probability")
    ax.set_ylabel("X-to-A Ratio = PiX / PiA")
    ax.axhline(y=expected_ratio, color='green', linestyle='--',linewidth=1, label='Expected Ratio')
    ax.legend()
    mp.show()



```
