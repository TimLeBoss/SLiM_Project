```python
# Star Migration
import matplotlib.pyplot as mp
import numpy as np
import os 
FOLDER = "/mnt/primevo/work/timothe_dandoy/SLiM_Project/tmp/Test_Migrations_NWFV/"
expected_ratio = 0.75

OF_Circle = []
OF_Star = []
OM_Circle = []
OM_Star = []
WMF_Circle = []
WMF_Star = []

for NUM in range(1,501):
    File = FOLDER+"NWFV_Mig_circle_Onlymale_"+str(NUM)+".txt"
    if os.path.isfile(File):
        Text = open(File, 'r')
        Values = Text.read().split()
        if len(Values) > 0:
            OM_Circle.append(float(Values[1])/float(Values[0]))
        Text.close()
    File = FOLDER+"NWFV_Mig_star_Onlymale_"+str(NUM)+".txt"
    if os.path.isfile(File):
        Text = open(File, 'r')
        Values = Text.read().split()
        if len(Values) > 0:
            OM_Star.append(float(Values[1])/float(Values[0]))
        Text.close()
    File = FOLDER+"NWFV_Mig_circle_Onlyfemale_"+str(NUM)+".txt"
    if os.path.isfile(File):
        Text = open(File, 'r')
        Values = Text.read().split()
        if len(Values) > 0:
            OF_Circle.append(float(Values[1])/float(Values[0]))
        Text.close()
    File = FOLDER+"NWFV_Mig_star_Onlyfemale_"+str(NUM)+".txt"
    if os.path.isfile(File):
        Text = open(File, 'r')
        Values = Text.read().split()
        if len(Values) > 0:
            OF_Star.append(float(Values[1])/float(Values[0]))
        Text.close()
    File = FOLDER+"NWFV_Mig_circle_wfemales_"+str(NUM)+".txt"
    if os.path.isfile(File):
        Text = open(File, 'r')
        Values = Text.read().split()
        if len(Values) > 0:
            WMF_Circle.append(float(Values[1])/float(Values[0]))
        Text.close()
    File = FOLDER+"NWFV_Mig_circle_wfemales_"+str(NUM)+".txt"
    if os.path.isfile(File):
        Text = open(File, 'r')
        Values = Text.read().split()
        if len(Values) > 0:
            WMF_Star.append(float(Values[1])/float(Values[0]))
        Text.close()
        
OF_Circle_median = np.median(OF_Circle)
OF_Star_median = np.median(OF_Star)
OM_Circle_median = np.median(OM_Circle)
OM_Star_median = np.median(OM_Star)
WMF_Circle_median = np.median(WMF_Circle)
WMF_Star_median = np.median(WMF_Star)

print("------------------------------------------------------------------------------------------------------------")
print("Migration Only female circle")
print("Number of simulations: ",len(OF_Circle))
print("Pi median Ratio: ",OF_Circle_median)
print("------------------------------------------------------------------------------------------------------------")
print("Migration Only female star")
print("Number of simulations: ",len(OF_Star))
print("Pi median Ratio: ",OF_Star_median)
print("------------------------------------------------------------------------------------------------------------")
print("Migration Only male circle")
print("Number of simulations: ",len(OM_Circle))
print("Pi median Ratio: ",OM_Circle_median)
print("------------------------------------------------------------------------------------------------------------")
print("Migration Only male Star")
print("Number of simulations: ",len(OM_Star))
print("Pi median Ratio: ",OM_Star_median)
print("------------------------------------------------------------------------------------------------------------")
print("Migration With female circle")
print("Number of simulations: ",len(WMF_Circle))
print("Pi median Ratio: ",WMF_Circle_median)
print("------------------------------------------------------------------------------------------------------------")
print("Migration With female Star")
print("Number of simulations: ",len(WMF_Star))
print("Pi median Ratio: ",WMF_Star_median)



print("------------------------------------------------------------------------------------------------------------")
fig1, ax1 = mp.subplots(figsize=(10, 5))
ax1.boxplot([OM_Star,OF_Star,WMF_Star],tick_labels=['OM_Star','OF_Star','WMF_Star'])
ax1.set_title("Star_Migration")
ax1.set_ylabel("X-to-A Ratio = PiX / PiA")
ax1.axhline(y=expected_ratio, color='green', linestyle='--',linewidth=1, label='Expected Ratio')
ax1.legend()
mp.show()

fig2, ax2 = mp.subplots(figsize=(10, 5))
ax2.boxplot([OM_Circle,OF_Circle,WMF_Circle],tick_labels=['OM_Circle','OF_Circle','WMF_Circle'])
ax1.set_title("Circle_Migration")
ax2.set_ylabel("X-to-A Ratio = PiX / PiA")
ax2.axhline(y=expected_ratio, color='green', linestyle='--',linewidth=1, label='Expected Ratio')
ax2.legend()
mp.show()
```
